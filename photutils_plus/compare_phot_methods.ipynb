```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import pylab
from astropy.io import ascii
```

```python
loc = '/Users/cgosmeyer/Desktop/test_cte/test_photom/phot_comparision/'

# Need to read data for the following scenarios:
# 1. IRAF coords + IRAF phot
# 2. IRAF coords + Python phot
# 3. Python coords + Python phot.
# all using params as identically as possible

# The data should be flux amount for each aperture. 

# This one will need special treatment?
data_irafc_irafp = ascii.read(loc+"id1l02g2q_flt.clean_cnts.fits0.mag.1")
data_irafc_pythonp = ascii.read(loc+"id1l02g2q_flt.clean.fits.mag")
data_pythonc_pythonp = ascii.read(loc+"id1l02g2q_flt.clean.fits.mag.1")

print data_irafc_irafp
print data_irafc_pythonp
print data_pythonc_pythonp
```

```python
def plot_flux_vs_aperture():
    """
    """
    #((flux[cutrangeA]/expt[cutrangeA])/divisorA - 1.0)*100.
    # calculate each as percent of the IRAF-IRAF data.
    
    radius = np.array(data_irafc_irafp['radius'])
    flux_irafc_irafp = np.array(data_irafc_irafp['aperture_sum'])
    flux_irafc_pythonp = np.array(data_irafc_pythonp['aperture_sum'])
    flux_pythonc_pythonp = np.array(data_pythonc_pythonp['aperture_sum'])

    
    fig=pylab.figure(figsize=(14.5,10.5))
    pylab.plot(radius, flux_irafc_irafp, \
               color='g', linewidth=3, label='IRAF coords, IRAF phot')
    pylab.plot(radius, flux_irafc_pythonp, \
               color='b', linewidth=3, label='IRAF coords, Python phot')
    pylab.plot(radius, flux_pythonc_pythonp, \
               color='orange', linewidth=3, label='Python coords, Python phot')

    pylab.xlabel('Aperture Radius [pixels]', fontsize=22, weight='bold')
    pylab.ylabel('Flux Counts', fontsize=22, weight='bold') 
    pylab.tick_params(axis='both', which='major', labelsize=20)
    pylab.title('Target: GRW70, Filter: F218W', fontsize=22)
    pylab.axis('tight')
    pylab.legend(loc='lower right')
    pylab.minorticks_on()
    pylab.savefig(loc+'compare_phot_methods1.png')    
  
    
    fig=pylab.figure(figsize=(14.5,10.5))
    pylab.plot(radius, (flux_irafc_pythonp/flux_irafc_irafp - 1.0)*100., \
               color='b', linewidth=3, label='IRAF coords, Python phot')
    pylab.plot(radius, (flux_pythonc_pythonp/flux_irafc_irafp - 1.0)*100., \
               color='orange', linewidth=3, label='Python coords, Python phot')

    pylab.xlabel('Aperture Radius [pixels]', fontsize=22, weight='bold')
    pylab.ylabel('% Flux difference from IRAF coords, IRAF phot', fontsize=22, weight='bold') 
    pylab.tick_params(axis='both', which='major', labelsize=20)
    pylab.title('Target: GRW70, Filter: F218W', fontsize=22)
    pylab.axis('tight')
    pylab.legend(loc='lower right')
    pylab.minorticks_on()
    pylab.xlim(0, 30)
    pylab.ylim(-5,1)
    pylab.savefig(loc+'compare_phot_methods2.png')  
    
plot_flux_vs_aperture()
```
